```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
# import warnings
import chardet

from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
df = df.replace('�', '', regex=True)
df
```

### data set cleaaning

```python
df = df.drop(columns=[
    'Restaurant Name','Address','Locality','Locality Verbose',
    'Currency','Rating color','Rating text',
    'Switch to order menu','Longitude','Latitude','Cuisines','City', 'Restaurant ID','Country Code'
])
```

```python
df.isnull().sum()
```

```python
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes':1,'No':0})
df['Has Table booking'] = df['Has Table booking'].map({'Yes':1,'No':0})
df['Is delivering now'] = df['Is delivering now'].map({'Yes':1,'No':0})
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
numeric_df
```

```python
numeric_df.corr()['Aggregate rating'].sort_values(ascending=False)
```

```python
df
```

### Train test split

```python
from sklearn.model_selection import train_test_split
df.columns
```

```python
input_columns = ['Average Cost for two',
       'Has Table booking', 'Has Online delivery', 'Is delivering now',
       'Price range', 'Votes']
output_columns = ['Aggregate rating']

x = df[input_columns]
y = df[output_columns]
split_test_size = 0.2


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = split_test_size,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")

x_train = imputer.fit_transform(x_train)
x_test = imputer.fit_transform(x_test)

```

```python
x_train[0:3]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test =scaler.transform(x_test)
```

### check the proportion

```python
print('{0:0.2f}% in training set'.format((len(x_train)/len(df.index))*100))


print('{0:0.2f}% in testing set'.format((len(x_test)/len(df.index))*100))
```

```python
df['Aggregate rating'].describe()
```

```python
print("Train mean:", y_train.mean())
print("Test mean :", y_test.mean())

print("Train min/max:", y_train.min(), y_train.max())
print("Test min/max :", y_test.min(), y_test.max())
```

## Train  the model using linear Regression

```python
lr_model= LinearRegression()
lr_model
```

```python
# training the model
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
y_pred = y_pred.ravel()
y_pred
```

```python
# R squared error
error_score = metrics.r2_score(y_test, y_pred)
print("R - squared error : ",error_score)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

## decision tree regression

```python
dtr_model = DecisionTreeRegressor()
dtr_model
```

```python
# training the model
dtr_model.fit(x_train,y_train)
```

```python
y_train_pred = dtr_model.predict(x_train)
y_test_pred = dtr_model.predict(x_test)
y_test_pred = y_test_pred.ravel()
y_test_pred
```

```python
# R squared error
error_score = metrics.r2_score(y_test, y_test_pred)
print("R - squared error : ",error_score)
```

```python
mse = mean_squared_error(y_test, y_test_pred)
print("Mean Squared Error:", mse)
```

## Conclusion

The Decision Tree Regression model performed better than Linear Regression.
Food quality and service ratings were the most influential factors affecting
restaurant aggregate ratings.
